```python
import numpy as np
import math
def k(X1, X2, s=1., ell=0.1):
    K = np.zeros((X1.shape[0], X2.shape[0]))
    for i in xrange(X1.shape[0]):
        for j in xrange(X2.shape[0]):
            tmp = (X1[i, :] - X2[j, :]) / ell
            K[i, j] = s ** 2 * math.exp(-.5 * np.dot(tmp, tmp))
    return K
```

```python
Xd = np.linspace(0, 1, 20)[:, None]
K = k(Xd, Xd)
import scipy.linalg
w, v = scipy.linalg.eigh(K)
w = w[::-1]
v = v[:, ::-1]
import matplotlib.pyplot as plt
j = 5
plt.plot(Xd, v[:, j], 'bo')
Xdd = np.linspace(0, 1, 100)[:, None]
Kdd = k(Xd, Xdd)
phi = np.dot(v.T, Kdd) / w[:, None]
plt.plot(Xdd, phi[j, :])
```

```python
for i in xrange(2):
    omega = np.random.randn(Xd.shape[0], 1) * np.sqrt(w[:, None] /  Xd.shape[0])
    s = np.dot(omega.T, phi)
    plt.plot(Xdd, s.T, 'b')
```
